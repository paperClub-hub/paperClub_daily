***
python文件类型主要如下：
|  python文件类型               | 说明                                           |
|  ----                     | ----              |
|embeddable zip file       | 是压缩包版本，即便携版，解压可用                     |
|executable installer      | 是可执行的安装版本，即离线版，下载到本地后可以直接安装 |
|web-based installer       |是联网安装版，体积很小，但需要保持网络畅通|
